```python
# decrypt .cram.cip files downloaded from EGA

import os,sys
temp_list = os.listdir("/home/users/team_projects/public_youk/Colon_Henry/EGAD00001004192/EGAD00001004192")
file_list = []
pre_list = []
for i in temp_list:
    if 'cram.cip' in i and 'md5' not in i:
        pre_list.append(i)
    if 'cram' in i and 'cram.cip' not in i and 'md5' not in i:
        file_list.append(i)
print len(pre_list)
print len(file_list)
#for i in pre_list:
#    print i
#sys.exit(1)
output_file=file("/home/users/team_projects/public_youk/Colon_Henry/EGAD00001004192/EGAD00001004192/00_decrypt.sh","w")
for input_fn in pre_list:
    if input_fn.replace(".cip","") not in file_list:
        output_file.write("openssl enc -d -aes-256-cbc -in %s -out %s -k Oc8lQCl529VoKYue2kHUcFnvKk5DSbaz -md sha256\n" %(input_fn,input_fn.replace(".cip","")))
output_file.close()
print 'THE END'
```

```python
#samtools merge

input_file = file("/home/users/team_projects/public_youk/Colon_Henry/EGAD00001004192/EGAD00001004192/xmls/EGAD00001004192_metadata.txt")
input_line = input_file.readline().strip()
input_line = input_file.readline().strip()

pd_id = []
file_name = []
output_file = file("/home/users/team_projects/public_youk/Colon_Henry/EGAD00001004192/EGAD00001004192/578_cram_files/00_make_578.sh","w")
while input_line:
    input_split = input_line.split('\t')
    if input_split[0] not in pd_id:
        pd_id.append(input_split[0])
        file_name.append([input_split[1]])
    else:
        file_name[pd_id.index(input_split[0])].append(input_split[1])
    input_line = input_file.readline().strip()
#print pd_id[0:10]
for i in range(0,len(pd_id)):
    if len(file_name[i]) == 1:
        output_file.write("cp ../%s %s\n" %(file_name[i][0],pd_id[i] + '.cram'))        
    else:
        temp = ''
        for j in file_name[i]:
            temp = temp + ' ../' + j
        output_file.write("samtools merge %s%s &> %s.out\n" %(pd_id[i]+'.cram', temp, pd_id[i]) )
output_file.close()
print 'THE END'
```
